```python
import pandas as pd
import numpy as numpy
import json          
```

```python
weapons=pd.read_excel('./assets.xlsx',sheet_name='weapon')
ids=weapons.loc[:,'id'].unique()
key0s=weapons.loc[:,0].unique()
key1s=weapons.loc[:,1].unique()
key2s=weapons.loc[:,2].unique()  
weaponJson={}
for id in ids:
    weaponJson[id]={}
    for key0 in key0s:
        weaponJson[id][key0]={}
        for key1 in key1s:
            weaponJson[id][key0][key1]={}
            for key2 in key2s:
                weaponJson[id][key0][key1][key2]={}  
weaponJsonFile=weapons.to_json(orient='index')
weaponJsonFile=json.loads(weaponJsonFile)

for e in weaponJsonFile:
    i=weaponJsonFile[e]
    weaponJson[i['id']][i['0']][i['1']][i['2']]=i

weaponJson=json.dumps(weaponJson)
f=open('./weapon.json','w')
f.write(weaponJson)
f.close()
```

```python
skills=pd.read_excel('./assets.xlsx',sheet_name='skill')
skillJsonFile=skills.to_json(orient='index')
skillJsonFile=json.loads(skillJsonFile)
 
skillJson={}
ids=skills.loc[:,'id'].unique()
types=skills.loc[:,'type'].unique()
sers=skills.loc[:,'ser'].unique()

for i in ids:
    skillJson[i]={}    
    for type in types:
        if(type=='A'):
            skillJson[i][type]={
                'pugong':{},
                'zhanji':{}
            }
        else:
            skillJson[i][type]={
                'zuijia':{},
                'dazhao':{}
            }


for e in skillJsonFile:
    i=skillJsonFile[e]
    skillJson[i['id']][i['type']][i['ser']]=i

skillJson=json.dumps(skillJson)
f=open('./skill.json','w')
f.write(skillJson)
f.close()
```
